```python
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

chromedriver = '/Users/jakekong/documents/chromedriver/chromedriver_mac64/chromedriver'
driver = webdriver.Chrome(chromedriver)

driver.get('http://datalab.naver.com/keyword/realtimeList.naver?where=main')
time.sleep(3)


date_withtime = driver.find_elements_by_css_selector('#content > div.keyword_carousel > div > div > div > div > strong')


hotwords_by_time = pd.DataFrame()     #for 문을 통해 나온 값들을 매번 append 하기 때문에, for 문 밖에 존재해야 함


#밑의 실험은, 각 date_withtime 이라는 컬럼이 될 변수에, 각각 따라올 실시간 검색어들이 바르게 관계를 맞기 위해 한번의 for 문으로
# hot_words 들을 각각의 date_withtime이 포괄 하도록 하였다.
for number in range(4):
    
    lst_key_words = []     #for 문이 시작하는 곳에 넣어주면, 매번 시작할때마다 다시 리스트를 비워줌. (for 문 밖에 있을시, 리스트에 추가식이 됨)
    
    #print (date_withtime[number].text) ------ 코딩이 제대로 되는지 위한 확인용
    a = date_withtime[number].text
    string_a = str(a)
    
    hot_words = driver.find_elements_by_css_selector('#content > div.keyword_carousel > div > div > div:nth-child({}) > div > div > ul > li > a > span'.format(number+1))
    
    for hot_word in hot_words:
        #print (hot_word.text)   -------- 코딩이 제대로 되는지 위한 확인용
        
        lst_key_words.append(hot_word.text)
    
    hotwords_by_time[string_a] = lst_key_words
    
driver.quit()

hotwords_by_time
```
