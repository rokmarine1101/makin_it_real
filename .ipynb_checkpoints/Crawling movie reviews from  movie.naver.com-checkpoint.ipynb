```python
import re
from bs4 import BeautifulSoup
import requests
import time


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains

res = requests.get('http://movie.naver.com/movie/running/current.nhn')   #사이트로 이동
soup = BeautifulSoup(res.content, 'html.parser')                        #html parser를 사용할 것을 명령



links = soup.select('dt > a[href]')      #parser를 이용해 dt 후 a에 href가 있는 모든 곳을 저장



for link in links[:5]:        #위의 명령으로 저장된 여러개의 location에서 하나하나씩 살펴봄
    back_link = link['href']      #해당사이트를 이동하기 위해 html에 저장되어 있는 링크를 저장; 그 이유는 링크가 http://로 시작하지 않음
    chromedriver = '/Users/jakekong/documents/chromedriver/chromedriver_mac64/chromedriver'
    driver = webdriver.Chrome(chromedriver)
    driver.get('http://movie.naver.com{}'.format(back_link))
    titles = driver.find_element_by_xpath("//*[@id='content']/div[1]/div[2]/div[1]/h3/a[1]")
    print('[',titles.text,']')
    
    
    try:
        element = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.ID, "content")))
                    
        more_views = driver.find_element_by_xpath("//*[@id='content']/div/div[4]/div[5]/div[2]/div/a")
                    
        webdriver.ActionChains(driver).click(more_views).perform()
    except:
        loop = False
    
    
    driver.switch_to_frame("pointAfterListIframe")  #iframe 데이터를 구하기 위한 작업
    
    try:
        for i in range(1,11):
            if i == 1:
                comments = driver.find_elements_by_xpath('/html/body/div/div/div[6]/ul/li/div/p')
                for comment in range(len(comments)):
                    print(comment+1,']', comments[comment].text)
            else:
                pages = driver.find_element_by_id("pagerTagAnchor{}".format(i))       
                pages.click()
                comments = driver.find_elements_by_xpath('/html/body/div/div/div[6]/ul/li/div/p')
                for comment in range(len(comments)):
                    print((i-1)*10+(comment+1),']',comments[comment].text)
    except:
        continue
        
    print ()
    print ()
        
                
    
    driver.quit()
```
