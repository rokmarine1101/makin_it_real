```python
import re
from bs4 import BeautifulSoup
import requests


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

chromedriver = '/Users/jakekong/documents/chromedriver/chromedriver_mac64/chromedriver' #내 맥북에 chromedriver 저장장소
driver = webdriver.Chrome(chromedriver)



res = requests.get("http://media.daum.net/economic/")
soup = BeautifulSoup(res.content, "html.parser")

links = soup.select("strong > a[href]")

for link in links:
    if re.search(r'^http://v.media.daum.net/v/\d+', link['href']):
        res2 = requests.get(link['href'])
        soup2 = BeautifulSoup(res2.content, "html.parser")
        title = soup2.find('title')
        driver.get(link['href'])
        

        
    
        for i in range(4):    #10개의 댓글을 확실하게 확보하기 위해 '더보기' 클릭킹을 어느정도 하게 만듬.
            try:
                element = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.ID, "alex-area")))
                    #xpath를 사용 하여 더보기 부분을 찾음.
                more_comments = driver.find_element_by_xpath("//*[@id='alex-area']/div/div/div/div[3]/div[1]/a")
                    
                    #위에 찾은 이미 명시된 '더보기'로 다가가서 클릭킹을 실시함 (perform) 
                webdriver.ActionChains(driver).move_to_element(more_comments).click(more_comments).perform() 
            except:
                loop = False
            

        comments = driver.find_elements_by_css_selector('#alex-area > div > div > div > div.cmt_box > ul > li > div.cmt_info > p')
        
        if re.findall(r'^.+뉴스$', title.get_text()):  #정규표현식을 통해 다음뉴스; 다음 연예 or 스포츠가 아닌; 만 추출 
            print('[',title.get_text(),']')           # [제목] 으로 나타냄
        
        
            if len(comments) > 10:                    #위의 충분한 더보기 클릭킹 후 댓글의 수가 10개 보다 많으면 10개 추출
                for ten_comments in range(0,10):
                    print (ten_comments+1,']',comments[ten_comments].text)    # 숫자] 댓글 으로 나타냄
        
        
            if len(comments) <10:                     #총 댓글의 수가 10개 보다 작을 시, 댓글의 갯수만큼 추출
                for few_comments in range(len(comments)):
                    print (few_comments+1,']',comments[few_comments].text)
        
            print ()
            print ()  #다음 링크의 제목과 댓글들을 육안적으로 보기 쉽게 하기 위함. 
                
    
    
driver.quit()
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

for_dataframe = []   #마지막에 각각의 리스트화 된 하나의 정보들의 집합을 저장시킬 리스트 변수를 만들어 준다.

locations = soup.find_all('location')

for location in locations:
    cities = location.find('city')
    data_collections = location.find_all('data')
    for data_collection in data_collections:
        date_time = data_collection.find('tmef')
        weather = data_collection.find('wf')
        for i in date_time:
            time = i.split()    # date_time의 문자들을 간격이 있는 곳으로 부터 split 해서 리스트 element화 시켜준다.
            if time[1] == '00:00': # time은 두개의 엘레멘트를 가지고 있고, 인덱스 0은 날짜, 인덱스 1은 시간으로 되어있다.
                time[1] = '오전'
            else:
                time[1] = '오후'
            for_dataframe.append ([cities.text, time[0], time[1], weather.text])

category = ['도시', '날짜', '시간', '날씨']

dataframe = pd.DataFrame(for_dataframe, columns = category)

dataframe
```

에 나오는 날짜별 실시간 검색어를 수집해서, 출력하세요 (dataframe으로 출력하셔도 됩니다.)

```python
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

chromedriver = '/Users/jakekong/documents/chromedriver/chromedriver_mac64/chromedriver'
driver = webdriver.Chrome(chromedriver)

driver.get('http://datalab.naver.com/keyword/realtimeList.naver?where=main')
time.sleep(3)


date_withtime = driver.find_elements_by_css_selector('#content > div.keyword_carousel > div > div > div > div > strong')


hotwords_by_time = pd.DataFrame()     #for 문을 통해 나온 값들을 매번 append 하기 때문에, for 문 밖에 존재해야 함


#밑의 실험은, 각 date_withtime 이라는 컬럼이 될 변수에, 각각 따라올 실시간 검색어들이 바르게 관계를 맞기 위해 한번의 for 문으로
# hot_words 들을 각각의 date_withtime이 포괄 하도록 하였다.
for number in range(4):
    
    lst_key_words = []     #for 문이 시작하는 곳에 넣어주면, 매번 시작할때마다 다시 리스트를 비워줌. (for 문 밖에 있을시, 리스트에 추가식이 됨)
    
    #print (date_withtime[number].text) ------ 코딩이 제대로 되는지 위한 확인용
    a = date_withtime[number].text
    string_a = str(a)
    
    hot_words = driver.find_elements_by_css_selector('#content > div.keyword_carousel > div > div > div:nth-child({}) > div > div > ul > li > a > span'.format(number+1))
    
    for hot_word in hot_words:
        #print (hot_word.text)   -------- 코딩이 제대로 되는지 위한 확인용
        
        lst_key_words.append(hot_word.text)
    
    hotwords_by_time[string_a] = lst_key_words
    
driver.quit()

hotwords_by_time
```
