```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

client_key = ''#클라이언트 키 는 동적으로 구할 것
area_code = '41590'
date = '201709'
#공공데이터 에서 개발계정, 서비스정보에서 end point 란에서 제공하는 주소를 가져옴
url = 'http://openapi.molit.go.kr:8081/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTrade?LAWD_CD={}&DEAL_YMD={}&serviceKey={}'.format(area_code, date, client_key)

information = []

response = requests.get(url)
soup = BeautifulSoup(response.content.decode('utf-8'), 'xml')
item_sets = soup.find_all('item')


for item_set in item_sets:
    price = item_set.find('거래금액').text
    year = item_set.find('년').text
    town = item_set.find('법정동').text
    apartment = item_set.find('아파트').text
    month = item_set.find('월').text
    day = item_set.find('일').text
    area = item_set.find('전용면적').text
    area_number = item_set.find('지번').text
    area_code = item_set.find('지역코드').text
    floor = item_set.find('층').text
    
    
    information.append([price,year,town,apartment,month,day,area,area_number,area_code,floor])

#information
info_columns = ['거래금액', '년', '법정동', '아파트', '월', '일', '전용면적', '지번', '지역코드', '층']

paper = pd.DataFrame(information, columns = info_columns)

paper
```
