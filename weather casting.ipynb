```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

for_dataframe = []   #마지막에 각각의 리스트화 된 하나의 정보들의 집합을 저장시킬 리스트 변수를 만들어 준다.

locations = soup.find_all('location')

for location in locations:
    cities = location.find('city')
    data_collections = location.find_all('data')
    for data_collection in data_collections:
        date_time = data_collection.find('tmef')
        weather = data_collection.find('wf')
        for i in date_time:
            time = i.split()    # date_time의 문자들을 간격이 있는 곳으로 부터 split 해서 리스트 element화 시켜준다.
            if time[1] == '00:00': # time은 두개의 엘레멘트를 가지고 있고, 인덱스 0은 날짜, 인덱스 1은 시간으로 되어있다.
                time[1] = '오전'
            else:
                time[1] = '오후'
            for_dataframe.append ([cities.text, time[0], time[1], weather.text])

category = ['도시', '날짜', '시간', '날씨']

dataframe = pd.DataFrame(for_dataframe, columns = category)

dataframe
```
